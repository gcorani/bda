```python
%matplotlib inline
import numpy as np
import arviz as az
import scipy.stats as stats
from matplotlib import pyplot as plt
az.style.use('arviz-darkgrid')
```

```python
x, y = np.mgrid[-4:4:.01, -4:4:.01]
pos = np.empty(x.shape + (2,))
pos[:, :, 0] = x; pos[:, :, 1] = y
rv = stats.multivariate_normal([0, 0], [[1, 0.5], [0.5, 1]])

x_value = pos[:, :, 0][:,0]
x_density = rv.pdf(pos)

# definitions for the axes
left, width = 0.1, 0.65
bottom, height = 0.1, 0.65
bottom_h = left_h = left + width + 0.02

rect_scatter = [left, bottom, width, height]
rect_histx = [left, bottom_h, width, 0.2]
rect_histy = [left_h, bottom, 0.2, height]

# start with a rectangular Figure
plt.figure(1, figsize=(8, 8))

ax_joint = plt.axes(rect_scatter)
ax_x = plt.axes(rect_histx)
ax_y = plt.axes(rect_histy)

# the scatter plot:
#ax_joint.contourf(x, y, rv.pdf(pos), cmap='Blues');
ax_joint.imshow(x_density, cmap='Blues', origin='lower', extent=[-3, 3, -3, 3])


ax_joint.plot(x_value, x_density[400]*2, 'k:', lw=2)
ax_joint.plot(x_value, x_density[500]*2+1, 'k:', lw=2)
ax_joint.plot(x_value, x_density[300]*2-1, 'k:', lw=2)

ax_x.plot(x_value, x_density.sum(1), 'k', lw=2)
ax_x.fill_between(x_value, x_density.sum(1), alpha=0.75, color='C0')
ax_y.plot(x_density.sum(0), x_value, 'k', lw=2)
ax_y.fill_betweenx(x_value, x_density.sum(1), alpha=0.75, color='C0')


for ax in [ax_joint, ax_x, ax_y]:
    ax.grid(False)
    ax.set_facecolor('w')
    ax.set_xticks([])
    ax.set_yticks([])
ax_joint.set_xlim(-3, 3)
ax_joint.set_ylim(-3, 3)
ax_x.set_xlim(-3, 3)
ax_y.set_ylim(-3, 3)
ax_x.set_xlim(-3, 3)
ax_joint.set_ylabel('y', rotation=0, labelpad=20, fontsize=18)
ax_joint.set_xlabel('x', fontsize=18)

ax_joint.text(-2.5, 2.5, 'p(x, y)', fontsize=18, color='k', weight='medium')
ax_y.text(10, 0, 'p(y)', fontsize=18, color='k', weight='medium')
ax_x.text(-0.2, 15, 'p(x)', fontsize=18, color='k', weight='medium')
ax_joint.text(1, -2, ' ... p(x | y)', fontsize=18, color='k', weight='medium')
plt.savefig('probabilidades.png', dpi=300)
```
