# Log score for categorical variables

* A  weatherperson issues probabilistic predictions for rain or shine on each day of the year.

* The predictions are in the form of probabilities of rain.

* Evaluate the forecast by computing the log-probability assigned to the correct events.

# Log-score for continous variables

\begin{align*}
\text{log-score: } &= \sum_i^n \log p(y_i)
\end{align*}
where:
* $y_1 ... y_n$: observations in the samples
* $p(y_i)$ density of the probabilistic prediction in correspondence of the actual value


# Example

* Consider a probabilitistic forecast with Gaussian distribution.

* Probabilistic forecast: $N(6,1^2)$: i.e.,  the expected value is 6, with standard deviation 1.
  * actual value: 5
  * log-score: evaluation of the normal density $N(6,1^2)$ in $x$=5.
  * $\text{pdfnorm}(x=5, loc=6, scale=1) = 0.24 $


* Probabilistic forecast: $N(7,1^2)$
  * actual value: 5
  * log-score: $ \text{pdfnorm}(x=5, loc=7, scale=1) = 0.05$
  * The log-score has worsened since the distance between expected and actual value has increased.
